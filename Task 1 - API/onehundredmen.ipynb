## Import requests
This is the module we will use to consume the API


```python
import requests
import json
```

## Define the URL
Create a variable, url, and set the url where the api sits. set the parameters using ?... use & to add another parameter.

```python
url = "https://randomuser.me/api/?results=100&gender=male"
```

## Obtain the data
Create a variable "response"and get the data using the request method found in the requests module.
The arguments are "GET" and url to show that we are obtaining (getting) our data from the url (the variable we defined).

```python
response = requests.request("GET", url)
```

print the response.

```python
print(response.text)
```

## Change to JSON format
Change to JSON format to work with it as a dictionary

```python
json_data = json.loads(response.text)
print(json_data)
```

## Extract needed data
We only want the user's name, email and gender. We extract this using a for loop. Remember that json_data is like a dictionary and within it we have "results" which is also like a dictionary. It's a nested dictionary. Within results we also have other dictionaries like "name".

```python
for data in json_data["results"]:
    first_name = data["name"]["first"]
    last_name = data["name"]["last"]
    gender = data["gender"]
    email = data["email"]

    print(first_name, " " + last_name, " " + gender, " " + email)

```
